# Changes in OpenTURNS from 1.13 to 1.14

The `ChangeLog` mentions that several OpenTURNS features changed :

https://github.com/openturns/openturns/blob/master/ChangeLog

The goal of this page is to share tricks to port old scripts to the newer versions of OpenTURNS.

## Empirical distribution function

OT 1.13:

    ot.VisualTest_DrawEmpiricalCDF(sample)
    
OT 1.14:

    ot.UserDefined(sample).drawCDF()

## Draw histogram

OT 1.12:

    ot.VisualTest_DrawHistogram(sample)

OT 1.13:

    ot.HistogramFactory().build(sample).drawPDF()

## RandomVector

OT 1.12:

    outputRandomVector = ot.RandomVector(g, inputRandomVector)

OT 1.13:

    outputRandomVector = ot.CompositeRandomVector(g, inputRandomVector)

## Gumbel

OT 1.13:

    myParam = ot.GumbelAB(1013.0, 558.0) # a = location, b = scale
    Q = ot.ParametrizedDistribution(myParam)

OT 1.14:

    Q = ot.Gumbel(558.0, 1013.0) # beta = scale, gamma = position

## Kolmogorov

OT 1.13:

    test_result = ot.FittingTest.Kolmogorov(sample, distributionFactory)

OT 1.14:

    distribution, test_result = ot.FittingTest.Kolmogorov(data, distributionFactory)

## LeastSquares

OT 1.13:

    responseSurface = myLeastSquares.getResponseSurface()

OT 1.14:

    responseSurface = myLeastSquares.getMetaModel()

## Beta

OT 1.13:

    r = 5
    t = 7
    a = 9
    b = 10
    x = ot.Beta(r, t, a, b)

OT 1.14:

    r = 5
    t = 7
    alpha = r
    beta = t - r
    a = 9
    b = 10
    x = ot.Beta(r, t, a, b)
