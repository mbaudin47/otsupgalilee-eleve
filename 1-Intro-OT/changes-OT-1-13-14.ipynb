# Changes in OpenTURNS from 1.13 to 1.14

This page gathers informations related to the changes from OpenTURNS 1.13 to 1.14.

* The `ChangeLog` mentions that several OpenTURNS features related to the parametrization of distributions changed in OpenTURNS 1.14:

https://github.com/openturns/openturns/blob/8eabdf99702a96e4bcfd2ccaeb04cc8620aa9208/ChangeLog#L40

* Several classes in `VisualTest` were removed:

https://github.com/openturns/openturns/blob/8eabdf99702a96e4bcfd2ccaeb04cc8620aa9208/ChangeLog#L93

One can use 

    import openturns.shims as ot
    
to maintain compatibility with older scripts. 

The goal of this page is to share tricks to port old scripts to the newer versions of OpenTURNS.

## Empirical distribution function

OT 1.13:

    ot.VisualTest_DrawEmpiricalCDF(sample)
    
OT 1.14:

    ot.UserDefined(sample).drawCDF()

## Draw histogram

OT 1.13:

    ot.VisualTest_DrawHistogram(sample)

OT 1.14:

    ot.HistogramFactory().build(sample).drawPDF()

## RandomVector

OT 1.13:

    outputRandomVector = ot.RandomVector(g, inputRandomVector)

OT 1.14:

    outputRandomVector = ot.CompositeRandomVector(g, inputRandomVector)

## Gumbel

OT 1.13:

    a = 1013.0
    b = 558.0
    myParam = ot.GumbelAB(a, b) # a = location, b = scale
    Q = ot.ParametrizedDistribution(myParam)

OT 1.14:

    a = 1013.0
    b = 558.0
    beta = b
    gamma = a
    Q = ot.Gumbel(beta, gamma) # beta = scale, gamma = position

## Kolmogorov

OT 1.13:

    test_result = ot.FittingTest.Kolmogorov(sample, distributionFactory)

OT 1.14:

    distribution, test_result = ot.FittingTest.Kolmogorov(data, distributionFactory)

## LeastSquares

OT 1.13:

    responseSurface = myLeastSquares.getResponseSurface()

OT 1.14:

    responseSurface = myLeastSquares.getMetaModel()

## Beta

OT 1.13:

    r = 5
    t = 7
    a = 9
    b = 10
    x = ot.Beta(r, t, a, b)

OT 1.14:

    r = 5
    t = 7
    alpha = r
    beta = t - r
    a = 9
    b = 10
    x = ot.Beta(alpha, beta, a, b)
