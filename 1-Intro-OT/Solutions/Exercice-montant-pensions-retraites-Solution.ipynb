# Modèle de pensions de retraites

```python
import openturns as ot
```

```python
import numpy as np
```

```python
first = 0.
width = [500,500,500,500,500,500,1000,1000]
height = [9,18,25,22,13,7,5,3]
dE = ot.Histogram(first, width, height)
dE.setDescription(["Ensemble"])
```

```python
first = 0.
width = [500,500,500,500,500,500,1000,1000]
height = [4, 11, 23, 25, 16, 9, 8, 5]
dH = ot.Histogram(first, width, height)
dH.setDescription(["Hommes"])
```

```python
first = 0.
width = [500,500,500,500,500,500,1000,1000]
height = [13, 24, 26, 19, 10, 5, 3, 1]
dF = ot.Histogram(first, width, height)
dF.setDescription(["Femmes"])
```

```python
graph_E = dE.drawPDF()
graph_E.setColors(["blue"])
graph_E
```

```python
graph_F = dF.drawPDF()
graph_F.setColors(["blue"])
graph_H = dH.drawPDF()
graph_H.setColors(["red"])
graph_F.add(graph_H)
graph_F
```

```python
print("Moyenne pension ensemble (EUR)")
muE = dE.getMean()[0]
muE
```

```python
print("Moyenne pension femmes (EUR)")
muF = dF.getMean()[0]
muF
```

```python
print("Moyenne pension hommes (EUR)")
muH = dH.getMean()[0]
muH
```

```python
alpha = 0.95
print("Intervalle à %.4f %% - Ensemble" % (alpha))
dE.computeBilateralConfidenceIntervalWithMarginalProbability(alpha)[0]
```

```python
print("Intervalle à %.4f %% - Femmes" % (alpha))
dF.computeBilateralConfidenceIntervalWithMarginalProbability(alpha)[0]
```

```python
print("Intervalle à %.4f %% - Hommes" % (alpha))
dH.computeBilateralConfidenceIntervalWithMarginalProbability(alpha)[0]
```

```python
sampleSize = 100
print("Taille de l'échantillon = %d" % (sampleSize))
```

```python
sampleE = dE.getSample(sampleSize)
meanE = sampleE.computeMean()
print("Pension moyenne de l'échantillon - Ensemble (EUR)")
meanE
```

```python
print("Pension moyenne exacte - Ensemble")
muE
```

```python
sampleF = dF.getSample(sampleSize)
meanF = sampleF.computeMean()
print("Pension moyenne de l'échantillon - Femmes (EUR)")
meanF
```

```python
print("Pension moyenne exacte - Femmes")
muF
```

```python
sampleH = dH.getSample(sampleSize)
meanH = sampleH.computeMean()
print("Pension moyenne de l'échantillon - Hommes (EUR)")
meanH
```

```python
print("Pension moyenne exacte - Hommes")
muH
```
